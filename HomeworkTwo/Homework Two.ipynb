### Functions for Question 1

```python
def plot_perceptron(line, weight_value, origin_value, line_color, \
                    other_line1 = None, line_color2 = None,\
                    other_line2 = None, line_color3 = None):
    x_coordinates = np.array([1, 3, 4])
    y_coordinates = np.array([3, 2, 1])
    x_values = np.linspace(-1, 6, 100)
    axes = plt.gca()
    axes.set_xlim([-5, 6])
    axes.set_ylim(-2, 6)
    plt.title("Class Points")
    plt.xlabel("x")
    plt.ylabel("y")
    
    plt.plot(x_values, line, line_color)
    plt.scatter(x_coordinates, y_coordinates, color=["purple", "blue", "green"])
    drawArrow(origin_value, weight_value)
    
    if line_color2 != None:
        plt.plot(x_values, other_line1, line_color2)
    if line_color3 != None:
        plt.plot(x_values, other_line2, line_color3)

def drawArrow(A, B):
    color = 'skyblue'
    plt.arrow(A[0], A[1], B[0] - A[0], B[1] - A[1], head_width=0.5, \
              head_length=0.5, fc=color, ec=color, length_includes_head = True)
```

## Functions For Question 2

```python
column_titles = ['frequency', 'angle_of_attack', 'chord_length', 'free-stream_velocity',\
                'suction_side_displacement_thickness', 'scaled_sound_pressure_level']

train_data = pd.read_csv('airfoil_self_noise_train.csv', names=column_titles)
test_data = pd.read_csv('airfoil_self_noise_test.csv', names=column_titles)

#Plots the histogram
def create_histogram(input_data, column_name):
    list_to_plot = input_data[column_name].tolist()
    #print(list_to_plot)
    size = len(set(list_to_plot))
    print(size)
    plt.title(column_name)
    plt.ylabel('frequency')
    plt.xlabel('value')
    plt.hist(list_to_plot, size, color="purple")
    plt.show()

    
#Accesses each column (except class) to call the histogram function
def loop_through_columns(input_data):
    for column_title in column_titles:
        create_histogram(input_data, column_title)

def loop_through_columns_2(input_data):
    for column_title in column_titles:
        if column_title != 'scaled_sound_pressure_level':
            removing_features(input_data, column_title)
        
def removing_features(input_data, column_name):
    input_data['intercept'] = pd.Series(np.array([1]*input_data.shape[0]),\
                                        index=input_data.index)
    input_data.drop(columns=[column_name])
    X = input_data.loc[:, input_data.columns != 'scaled_sound_pressure_level'].values
    y = input_data.scaled_sound_pressure_level.values
    weights = np.linalg.inv(X.T @ X) @ X.T @ y
    y_hat = X @ weights
    rss = (y - y_hat).T @ (y - y_hat)
    print("RSS = ", rss, " when ", column_name, " removed")
```
